## **반환값(return)**

함수를 실행하면 일종의 명령어처럼 작동합니다.  
그렇지만 함수는 작동하고 나면 자신이 끝났다는 신호를 반드시 보냅니다.  
그것을 반환값이라고 부릅니다.





#### **예제 3 : 반환값 실험**

```python
# 예제 3 : 반환값 실험

def greet():
    print("함수 실험")
    
def greet_num(n):
    print(n, "함수 실험2")
    
greet()
print("-----------------")
greet_num(10)
print("-----------------")
print(greet())
print("-----------------")
print(greet_num(10))
print("-----------------")
```

#### 예제 3 해설

> 함수 `greet()`는 "함수 실험"이라는 글을 출력하는 함수입니다.
> 함수 `greet_num(n)`은 숫자와 "함수 실험2"라는 글을 출력하는 함수입니다.
>
> 그런데 이것을 `print()`에 넣어서 실행해도 함수가 작동합니다.
>
> 이때 `None`(아무것도 없음)이라는 것이 같이 출력됩니다.
> 반환값이 없다는 뜻입니다.
>
> 이 과정은 정리하면 다음과 같습니다.
> 
> `print(명령어()) ` 
> 1. 명령어()를 실행
> 2. 명령어()가 끝나고 내놓는 반환값 출력
> 3. 반환값이 없으므로 `None` 출력
>   
> `greet()`나 `greet_num(n)`은 명령어가 끝나고 보내는 반환값이 없으므로 `None`이 출력됩니다.
>
> 반환값을 만들어주는 경우는 어떨까요? 다음 코드를 실행해봅시다.  

#### **예제 4 : 반환값 기초**

```python
# 예제 4 : 반환값 기초

def add_multiple(n, m):
    res = (n + m) * m
    return res

add_multiple(10, 20)
print("-----------------")
t = add_multiple(10, 20)
print(t)
print("-----------------")
print(add_multiple(10, 20))
```

#### 예제 4 해설

> `add_multiple(n, m)`은 `n`에 `m`만큼 더한 다음에 `m`을 곱하는 함수입니다.   
그렇지만 함수를 그냥 실행하면 아무 일도 일어나지 않는 것처럼 보입니다.  
함수 안에는 `print()`가 없기 때문이지요.
>
> `t = add_multiple(10, 20)`는
`add_multiple(10, 20)`의 반환값을 `t`에 저장하라는 뜻입니다.  
따라서 `t`에는 `(10+20)*20`의 결과인 `600`이 저장됩니다.
>
> `print(add_multiple(10, 20))`은 함수의 반환값인 `600`이 출력됩니다.
